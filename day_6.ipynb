```python
# game 2
# contact_book_application
contacts = {}
def add_contact(name, number,contacts):
    contacts[name] = number
    with open("contacts.txt", "a") as file:
        file.write(f"{name},{number}\n")
def remove_contact(name,contacts):
    if name in contacts:
        del contacts[name]
        with open("contacts.txt", "w") as file:
            for key, value in contacts.items():
                file.write(f"{key},{value}\n")
        print("Contact removed successfully")
    else:
        print("Contact not found")
def search_contact(name,contacts):
    if name in contacts:
        return contacts[name]
        with open("contacts.txt", "r") as file:
            for line in file:
                name, number = line.strip().split(",")
                if name == name:
                    return number
                else:
                    print("not found.")
    else:
        return None
def list_contacts(contacts):
    # return contacts.keys
    with open("contacts.txt", "r") as file:
        for line in file:
            name, number = line.strip().split(",")
            print(f"{name}: {number}")

while True:
        print("\n1. Add Contact\n2. Remove Contact\n3. Search Contact\n4. List Contacts\n5. Quit")
        choice = input("Enter your choice: ")
        if choice == '1':
            name = input("Enter the name of the contact: ")
            number = input("Enter the number of the contact: ")
            add_contact(name, number, contacts)
        elif choice == '2':
            name = input("Enter the name of the contact to remove: ")
            remove_contact(name, contacts)
        elif choice == '3':
            name = input("Enter the name of the contact to search: ")
            print(search_contact(name, contacts))
        elif choice == '4':
            list_contacts(contacts)
        elif choice == '5':
            break
        else:
            print("Invalid choice. Please try again.")
```

```python
import random

stages = [
      """
          --------
          |      |
          |
          |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |      |
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|\
          |
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|\
          |     /
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|\
          |     / \
          |
          |
          -------
      """,
      """
          --------
          |      |
          |      O
          |     /|\
          |     / \
          |    /
          |
          -------
      """
  ]

def generate_word():
  """Chooses a random word from a list, with the option to customize the word list."""
  words = ["apple", "banana", "orange", "computer", "python"]
  # Optionally, allow user-defined word list:
  custom_word = input("Enter a custom word (leave blank for random): ").lower()
  if custom_word:
    return custom_word
  else:
    return random.choice(words)

def display_hangman(incorrect_guesses):
  """Displays the hangman figure based on incorrect guesses."""

  return stages[incorrect_guesses]

def play_hangman():
  """Main function for the hangman game."""
  word = generate_word().lower()  # Choose a random (or custom) word and convert to lowercase
  guessed_letters = set()  # Set to store guessed letters (avoids duplicates)
  incorrect_guesses = 0

  print("Welcome to Hangman!")

  while incorrect_guesses < len(stages) - 1:
    # Display current word state with underscores for unguessed letters
    display_word = []
    for letter in word:
      if letter in guessed_letters:
        display_word.append(letter)
      else:
        display_word.append("_")
    print(" ".join(display_word))

    # Get player's guess
    guess = input("Guess a letter: ").lower()

    if guess in guessed_letters:
      print("You already guessed that letter.")
    elif guess in word:
      guessed_letters.add(guess)
    else:
      incorrect_guesses += 1
      print(f"Wrong guess! {display_hangman(incorrect_guesses)}")

    # Check if word is guessed or game over
    if all(letter in guessed_letters for letter in word):
      print("Congratulations! You guessed the word!")
      break

  if incorrect_guesses == len(stages) - 1:
    print(f"You ran out of guesses. The word was: {word}")
    print(display_hangman(incorrect_guesses))

if __name__ == "__main__":
  play_hangman()
```
